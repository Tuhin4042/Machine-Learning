## split features and target

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

<h2 style=color:blue>  If the code is working,then its provides the best accuracy</h2> 
<h3 style=color:red> Most of the time the compiler can't run it</h3>

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import numpy as np

tune = []
accuracies = []
kernel_Value = ['linear', 'poly', 'rbf', 'sigmoid']

for i in range(100):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=i)
    
    for k in kernel_Value:
        svm = SVC(kernel=k, C=30, gamma='auto')
        svm.fit(X_train, y_train)
        accuracies.append(svm.score(X_test, y_test) * 100)
        tune.append((i, k))

Accuracy = max(accuracies)
values = tune[np.argmax(accuracies)]

print(f"Best Accuracy: {Accuracy:.2f}%")
print(f"Best Random State: {values[0]}")
print(f"Best Kernel: {values[1]}")
```
